```python
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model= SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentence=["The cat sat on the mat.","A feline rested on the rug.","Quantum mechanics describes subatomic particles.",
"I love programming in Python.","Python snakes are common in Asia.","The Function in python is important."]
```

```python
embeding=model.encode(sentence)
```
